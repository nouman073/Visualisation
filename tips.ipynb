```python
# import some important libraries 
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#import dataset of tips
df=sns.load_dataset('tips')
```

```python
plt.style.use('fivethirtyeight')
```

## Strip Plot

It displays the distribution of a continuous variable across different categories.

Data points for each category are plotted along a single axis, with a small amount of random noise added to their positions.

Strip plots are less computationally intensive than swarm plots and can be useful for quick data exploration.

They can suffer from overplotting, which occurs when multiple data points have the same value and are plotted on top of each other.

Strip plots can be combined with other types of plots, such as box plots or violin plots, for more detailed analysis.

```python
sns.catplot(x='day',y='tip', kind='strip', data=df, hue='day')
```

-In this plot add 'jitter' parameter . that parameter use for add random noise to the positions of data points

 -The purpose of jitter is to avoid overplotting.

 -Jitter can help make the distribution of the data more visible and easier to interpret.

```python
sns.catplot(x='day',y='tip', kind='strip',jitter=1.5, data=df, hue='day')
```

# Swarm Plot

Swarm plots avoid overplotting by adding a random jitter to the position of each data point along the categorical axis.
They are useful for visualizing the distribution of a continuous variable across different categories.
The orientation of the plot can be changed to display the data vertically or horizontally.

```python
sns.catplot(x='day',y='tip',kind='swarm',hue='sex',data=df)
```
